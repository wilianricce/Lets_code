```python
# Visualizando a distribuição da variável Nota2
df_dados["Nota2"].plot.hist(bins=10)
```

```python
# Média e Mediana estão com valores próximos e, com base nos histogramas, é possível verificar 
# distribuição regular entre as classes. Assim, as falhas serão preenchidas com a média dos valores de cada coluna

df_dados = df_dados.fillna({'Nota1': df_dados['Nota1'].mean(), 'Nota2': df_dados['Nota2'].mean()})
```

```python
# Verificar o número de dados faltantes
print("Número de dados faltantes por variável, após processamento:")
df_dados.isna().sum()
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50


```python
# Criar coluna "Total" somando o valor de todos os ativos por investidor
df_dados['Total'] = df_dados['Ativo01'] + df_dados['Ativo02'] + df_dados['Ativo03'] + df_dados['Ativo04'] + df_dados['Ativo05'] + df_dados['Ativo06'] + df_dados['Ativo07']
```

```python
# Criar coluna "Media" com a média das notas 1 e 2
df_dados['Media'] = (df_dados['Nota1'] + df_dados['Nota2']) / 2
```

```python
# Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
df_dados['Dominio'] = df_dados['Email'].str.split('@', expand=True)[1]
```

```python
# Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)

df_dados.rename(columns={'Salario':'Renda'}, inplace=True)

df_dados['Renda'] = df_dados['Renda'].str.strip("$").astype('float64')
```

```python
# Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)
df_dados['Sexo'] = df_dados['Sexo'].map({'Masc': 'Masculino', 'Fem':'Feminino'})
```

```python
# Transformar os dados da coluna Media em inteiro
df_dados['Media'] = df_dados['Media'].astype(int)
```

```python
# Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, 
# como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50

df_dados.loc[df_dados["Media"] < 50, "NPS"] = "detrator"
df_dados.loc[(df_dados["Media"] >= 50) & (df_dados["Media"] < 80), "NPS"] = "neutro"
df_dados.loc[df_dados["Media"] >= 80, "NPS"] = "promotor"
```

```python
# Visualização do dataframe com as alterações propostas
df_dados.head()
```

- Qual o menor valor investido em cada ativo?
- Mostre a média e desvio padrão de cada ativo

```python
# Mostre a média e desvio padrão de cada ativo
print("Média dos valores invetidos por ativo: ")
df_dados.loc[:,'Ativo01':'Ativo07'].mean().apply(lambda x: "{:.2f}".format(x))
```

```python
print("Desvio padrão dos valores investidos em cada ativo: ")
df_dados.loc[:,'Ativo01':'Ativo07'].std().apply(lambda x: "{:.2f}".format(x))
```

```python
# Mostre os 10 primeiros ids dos investidores de perfil Agressivo e Moderado.
print("Perfil Agressivo:")
df_dados[df_dados['Perfil']=="Agressivo"]['id'][:10]
```

```python
print("Perfil Moderado:")
df_dados[df_dados['Perfil']=="Moderado"]['id'][:10]
```

```python
# Quantos investidores investiram mais de 16000 no Ativo04?
print("Investivores com mais de 16 mil no Ativo04:")
df_dados.loc[:,'Ativo04'][df_dados.loc[:,'Ativo04']>16000].count()
```

```python
# Quantos investidores investiram mais de 16000 no Ativo04 e menos de 5000 no Ativo05?
print("Investivores com mais de 16 mil no Ativo04 e menos de 5 mil no Ativo05:")
df_dados.loc[:,['Ativo04', 'Ativo05']][(df_dados.loc[:,'Ativo04']>16000) & (df_dados.loc[:,'Ativo05']<5000)].shape[0]
```

```python
# Quais investidores do sexo feminino tem mais de 25 anos e seu investimento total foi menor que 10000?
print("Investidores do sexo feminino tem mais de 25 anos e seu investimento total foi menor que 10000: ")
df_dados.loc[:,['Sexo', 'Total']][(df_dados.loc[:,'Sexo']=='Feminino') & (df_dados.loc[:,'Total']<10000)].shape[0]
```

```python
# Quantos investidores tem em cada estado (dica: use groupby)
print('Investidores por Estado (UF):')
df_dados.groupby('Estado').count()['id']
```

```python
# Qual é a soma do total investido por cada estado?
print('Soma total dos valores investido por Estado (UF:)')
df_dados.groupby('Estado')['Total'].sum()
```

```python
# Qual é a média do total investido por promotores, detratores e neutros? E como se dividem nos estados
print('Média investida por NPS:')
df_dados.groupby('NPS')['Total'].mean()
```

```python
# Como se dividem nos estados
print("Média investida por NPS e Estado (UF):")
df_dados.groupby(['Estado','NPS'])[['Total']].mean().unstack()
```

```python
# Crie 3 colunas "dummies" do perfil dos investidores
df_dados[['Agressivo','Conservador','Moderado']] = pd.get_dummies(df_dados['Perfil'])
df_dados.head()
```

```python
# Crie uma tabela com total de investimentos de promotores, detratores e neutros em função do perfil.
# (dica: use pivot_table)
pd.pivot_table(df_dados, values='Total', index=['NPS'], columns=['Perfil'], aggfunc=(np.sum)).style.format('{0:,.2f}')
```
